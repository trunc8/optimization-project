```python
final_val_obj=df.iloc[19][13]
final_val_obj
```

```python
#ready the bounds because they are required in objective function 1
l1 = [[20121,30180]]*4
l2 = [[640,960]]*4
l3 = [[2,3.5]]*2
l4 = [[0.75,1.2]]*2
l1.extend(l2)
l1.extend(l3)
l1.extend(l4)
bounds=l1
```

```python
series_of_values=[0 for i in range(13)]
```

```python
def evaluate(index,parameters,design_variables):
    f=objective_function_1(parameters,design_variables,bounds)
    #ans=del f/del variable
    ans=abs(final_val_obj-f)
    series_of_values[index]=ans
    
```

```python
# Now comes the real part
i=0
while i<13:
    #for each variable change its value by +1
    if (i==12):
        parameters[0]=df.iloc[19][i]+1
        evaluate(i,parameters,design_variables)
        parameters[0]=df.iloc[19][i] # restoring the previous value
        
    else:
        design_variables[i]=df.iloc[19][i]+1
        evaluate(i,parameters,design_variables)
        design_variables[i]=df.iloc[19][i]  # restoring the previous value
    i+=1
        
```
